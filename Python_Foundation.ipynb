# Meng-update DataFrame

```python
#Menghitung BMI setiap siswa
df_a['BMI'] = df_a['Berat Badan (kg)'] / ((df_a['Tinggi Badan (cm)'] / 100)**2)
```

```python
print("DataFrame yang sudah diupdate")
df_a
```

Mengkategorikan siswa berdasarkan BMI
* Normal: 18.5-24.9
* Overweight: 25.0-29.9
* Obesity: 30.0-34.9
* Underweight: < 18.5


```python
def get_status(BMI):
  if BMI < 18.5:
    return 'Underweight'
  elif 18.5 <= BMI <= 24.9:
    return 'Normal'
  elif 25.0 <= BMI <= 29.9:
    return 'Normal'
  elif 30.0 <= BMI <=34.9:
    return 'Obesity'
  else:
    return 'Extreme Obesity'

df_a['Status BMI'] = df_a['BMI'].apply(get_status)
df_a
```
